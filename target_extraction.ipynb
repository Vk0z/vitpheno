```python
import geopandas as gpd
import pandas as pd
```

```python
# Read the envelopes GeoDataFrame
envelopes_gdf = gpd.read_file(r"C:\Users\Kostas\Desktop\GIMA\Module_7\Data\PEP725\After_2016_sent_from_PEP725\pep725_outputs\PEP725_envelopes.geojson")
envelopes_gdf.set_crs(32632, inplace=True, allow_override=True)
envelopes_list = envelopes_gdf.geometry.tolist()
```

```python
# Best approach? Using agg method. 

# Group by 's_id' and 'date', 'lat', 'lon' are just to include in the df, and find the most frequent value for 'Label' and 'phase_id'
# agg goes through the entire columns of 'Label' and 'phase_id' and aggregates them based on the lambda function
# In this case it uses the value_counts() method and then takes the first result, which is the most frequent value
freqresults_df = envelopes_gdf.groupby(['s_id', 'date', 'lat', 'lon'])[['Label', 'phase_id']].agg(lambda x: x.value_counts().index[0])

# Reset the index to turn 's_id' and 'date' into regular columns
freqresults_df = freqresults_df.reset_index()
```

```python
# Converting the date column to datetime for sorting and subsequent operations
freqresults_df['date'] = pd.to_datetime(freqresults_df['date'], format='%Y-%m-%d').dt.date
```

```python
freqresults_df.sort_values(['s_id', 'date'], inplace=True)
```
